```python
from pdbfixer import PDBFixer
from openmm.app import PDBFile
```

```python
#fixer = PDBFixer(filename="binders/seq6_binder/protein_seq6_binder.pdb")
fixer = PDBFixer(filename="nonbinders/seq1_nb/protein_seq1_nb.pdb")

# Detect problems
fixer.findMissingResidues()
fixer.findMissingAtoms()

# Add missing heavy atoms
fixer.addMissingAtoms()

# Add hydrogens at desired pH
fixer.addMissingHydrogens(pH=7.0)

# # Write binder fixed structure
# with open("binders/seq6_binder/protein_seq6_binder_fixed_H.pdb", "w") as f:
#     PDBFile.writeFile(fixer.topology, fixer.positions, f)

# Write nonbinder fixed structure
with open("nonbinders/seq1_nb/protein_seq1_nb_fixed_H.pdb", "w") as f:
    PDBFile.writeFile(fixer.topology, fixer.positions, f)
```
