```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
# Load PDB
#mol = Chem.MolFromPDBFile("binders/seq6_binder/ligand_seq6_binder.pdb", removeHs=False) #binder
mol = Chem.MolFromPDBFile("nonbinders/seq1_nb/ligand_seq1_nb.pdb", removeHs=False) #nonbinder

# Add hydrogens (if missing)
mol = Chem.AddHs(mol)

# Assign bond orders + geometry
AllChem.EmbedMolecule(mol, AllChem.ETKDG())
AllChem.UFFOptimizeMolecule(mol)

# Write SDF
#w = Chem.SDWriter("binders/seq6_binder/ligand_seq6_binder.sdf") #binder
w = Chem.SDWriter("nonbinders/seq1_nb/ligand_seq1_nb.sdf") #nonbinder
w.write(mol)
w.close()
```
