```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
# -------------------------
# ----- User inputs -------
ID_list = ["0014", "0263", "0540", "0997", "1745"]

#seq_name = "seq"
seq_name = "pair"

#seq_type = "binder"
#seq_type = "nonbinder"
#seq_type = "neg_low_pkt"
seq_type = "neg_fail_gate"
# -------------------------

for i in range(len(ID_list)):
    ID = ID_list[i]
    # Load PDB
    if seq_type == "binder":
        if seq_name == "seq":
            mol = Chem.MolFromPDBFile(f"binders/seq{ID}_binder/ligand_seq{ID}_binder.pdb", removeHs=False)
        else:
            mol = Chem.MolFromPDBFile(f"binders/pair_{ID}_binder/ligand_pair{ID}.pdb", removeHs=False)
    elif seq_type == "nonbinder":
        if seq_name == "seq":
            mol = Chem.MolFromPDBFile(f"nonbinders/seq{ID}_nb/ligand_seq{ID}_nb.pdb", removeHs=False)
        else:
            mol = Chem.MolFromPDBFile(f"nonbinders/pair_{ID}_nb/ligand_pair{ID}.pdb", removeHs=False)
    elif seq_type == "neg_low_pkt":
        mol = Chem.MolFromPDBFile(f"neg_low_pkt/pair_{ID}_low_pkt/ligand_pair{ID}.pdb", removeHs=False)
    elif seq_type == "neg_fail_gate":
        mol = Chem.MolFromPDBFile(f"neg_fail_gate/pair_{ID}_fail_gate/ligand_pair{ID}.pdb", removeHs=False)
    
    # Write SDF
    if seq_type == "binder":
        #w = Chem.SDWriter(f"binders/seq{ID}_binder/ligand_seq{ID}_binder.sdf")
        w = Chem.SDWriter(f"binders/pair_{ID}_binder/ligand_pair{ID}.sdf")
    elif seq_type == "nonbinder":
        #w = Chem.SDWriter(f"nonbinders/seq{ID}_nb/ligand_seq{ID}_nb.sdf")
        w = Chem.SDWriter(f"nonbinders/pair_{ID}_nb/ligand_pair{ID}.sdf")
    elif seq_type == "neg_low_pkt":
        w = Chem.SDWriter(f"neg_low_pkt/pair_{ID}_low_pkt/ligand_pair{ID}.sdf")
    elif seq_type == "neg_fail_gate":
        w = Chem.SDWriter(f"neg_fail_gate/pair_{ID}_fail_gate/ligand_pair{ID}.sdf")
    
    w.write(mol)
    w.close()
```
