```python
all_pdb_df = pd.read_csv(f"orig_data/md_candidate_guide.csv")
#all_pdb_df = pd.read_csv(f"orig_data/md_condidate_open_gate_pdbs/open_gate_guide.csv")
all_pdb_df
```

```python
pair_ids = all_pdb_df.iloc[:,0]
md_group_names = all_pdb_df.iloc[:,1].unique()
md_group_names
```

```python
binder_pairIDs, nb_pairIDs, neg_low_pkt_pairIDs, neg_fail_gate_pairIDs = [], [], [], []
for i in range(len(all_pdb_df)):
    if all_pdb_df.iloc[:,1][i] == "binder":
        binder_pairIDs.append(all_pdb_df.iloc[:,0][i])
    elif all_pdb_df.iloc[:,1][i] == "non_binder":
        nb_pairIDs.append(all_pdb_df.iloc[:,0][i])
    elif all_pdb_df.iloc[:,1][i] == "negative_low_pocket":
        neg_low_pkt_pairIDs.append(all_pdb_df.iloc[:,0][i])
    elif all_pdb_df.iloc[:,1][i] == "negative_fail_gate":
        neg_fail_gate_pairIDs.append(all_pdb_df.iloc[:,0][i])

print("Num binders: ", len(binder_pairIDs))
print("Num nonbinders: ", len(nb_pairIDs))
print("Num negative_low_pocket: ", len(neg_low_pkt_pairIDs))
print("Num negative_fail_gate: ", len(neg_fail_gate_pairIDs))
print("Total: ", len(binder_pairIDs)+len(nb_pairIDs)+len(neg_low_pkt_pairIDs)+len(neg_fail_gate_pairIDs))

print("First 5 binders: ", binder_pairIDs[0:5])
print("First 5 nonbinders: ", nb_pairIDs[0:5])
print("First 5 negative_low_pocket: ", neg_low_pkt_pairIDs[0:5])
print("First 5 negative_fail_gate: ", neg_fail_gate_pairIDs[0:5])
```
